Organizing data in a diferent way

```python
ages = []
sexes = []
bmis = []
num_children = []
smoker_statuses = []
regions = []
insurance_charges = []


def load_list_data(lst, csv_file, column_name):
    # open csv file
    with open(csv_file) as csv_info:
        # read the data from the csv file
        csv_dict = csv.DictReader(csv_info)
        # loop through the data in each row of the csv 
        for row in csv_dict:
            # add the data from each row to a list
            lst.append(row[column_name])
        # return the list
        return lst
```

```python
load_list_data(ages, 'insurance.csv', 'age')
load_list_data(sexes, 'insurance.csv', 'sex')
load_list_data(bmis, 'insurance.csv', 'bmi')
load_list_data(num_children, 'insurance.csv', 'children')
load_list_data(smoker_statuses, 'insurance.csv', 'smoker')
load_list_data(regions, 'insurance.csv', 'region')
load_list_data(insurance_charges, 'insurance.csv', 'charges')
```

```python
class PatientsInfo:
    # init method that takes in each list parameter
    def __init__(self, patients_ages, patients_sexes, patients_bmis, patients_num_children, 
                 patients_smoker_statuses, patients_regions, patients_charges):
        self.patients_ages = patients_ages
        self.patients_sexes = patients_sexes
        self.patients_bmis = patients_bmis
        self.patients_num_children = patients_num_children
        self.patients_smoker_statuses = patients_smoker_statuses
        self.patients_regions = patients_regions
        self.patients_charges = patients_charges

    # method that calculates the average ages of the patients in insurance.csv
    def analyze_ages(self):
        # initialize total age at zero
        total_age = 0
        # iterate through all ages in the ages list
        for age in self.patients_ages:
            # sum of the total age
            total_age += int(age)
        # return total age divided by the length of the patient list
        return ("Average Patient Age: " + str(round(total_age/len(self.patients_ages), 2)) + " years")

    # method that calculates the number of males and females in insurance.csv
    def analyze_sexes(self):
        # initialize number of males and females to zero
        females = 0
        males = 0
        # iterate through each sex in the sexes list
        for sex in self.patients_sexes:
            # if female add to female variable
            if sex == 'female':
                females += 1
            # if male add to male variable
            elif sex == 'male':
                males += 1
        # print out the number of each
        print("Count for female: ", females)
        print("Count for male: ", males)

    # method to find each unique region patients are from
    def unique_regions(self):
        # initialize empty list
        unique_regions = []
        # iterate through each region in regions list
        for region in self.patients_regions:
            # if the region is not already in the unique regions list
            # then add it to the unique regions list
            if region not in unique_regions: 
                unique_regions.append(region)
        # return unique regions list
        return unique_regions

    # method to find average yearly medical charges for patients in insurance.csv
    def average_charges(self):
        # initialize total_charges variable
        total_charges = 0
        # iterate through charges in patients charges list
        # add each charge to total_charge
        for charge in self.patients_charges:
            total_charges += float(charge)
        # return the average charges rounded to the hundredths place
        return ("Average Yearly Medical Insurance Charges: " +  
                str(round(total_charges/len(self.patients_charges), 2)) + " dollars.")
    
    # method to create dictionary with all patients information
    def create_dictionary(self):
        self.patients_dictionary = {}
        self.patients_dictionary["age"] = [int(age) for age in self.patients_ages]
        self.patients_dictionary["sex"] = self.patients_sexes
        self.patients_dictionary["bmi"] = self.patients_bmis
        self.patients_dictionary["children"] = self.patients_num_children
        self.patients_dictionary["smoker"] = self.patients_smoker_statuses
        self.patients_dictionary["regions"] = self.patients_regions
        self.patients_dictionary["charges"] = self.patients_charges
        return self.patients_dictionary
```

-- There are differences between the actual charges and the estimated charges (insurance costs) - (based on previous US MEDICAL INSURANCE Projects)?

```python
# Converting categorical variables (sex and smoker) to binary ones.
# Male - 1; Female - 0; Smoker - 1; Non-smoker - 0

converted_sex = [1 if gender == "male" else 0 for gender in sexes]
print(converted_sex, "\n")

converted_smoker = [1 if status == "yes" else 0 for status in smoker_statuses ]
print(converted_smoker, "\n")
```

```python
# Preparing data to perform calculations

ages = np.array(ages, dtype=np.float64)
converted_sex = np.array(converted_sex, dtype=np.float64)
bmis = np.array(bmis, dtype=np.float64)
num_children = np.array(num_children, dtype=np.float64)
converted_smoker = np.array(converted_smoker, dtype=np.float64)
insurance_charges = np.array(insurance_charges, dtype=np.float64)

print(ages.dtype)
print(converted_sex.dtype)
print(bmis.dtype)
print(num_children.dtype)
print(converted_smoker.dtype)
print(insurance_charges.dtype) 
```

```python
# Creating a list of estimated cost for each insurance charge on medical data
estimated_cost = []

for i in range(len(insurance_charges)):
  cost = round((250 * ages[i]) - (128 * converted_sex[i]) + (370 * bmis[i]) + (425 * num_children[i]) + (24000 * converted_smoker[i]), 2)
  estimated_cost.append(cost)
print(estimated_cost)

# Add the estimated insurance cost to our medical data

medical_data['Estimated_cost'] = estimated_cost
print(medical_data)
```

```python
# Verifying de diferences between actual charges and estimated charges  

dif_charges = medical_data['charges'] - medical_data['Estimated_cost']
print(dif_charges, "\n")

average_diff_charges = dif_charges.sum()/len(medical_data['Estimated_cost'])
print(round(average_diff_charges, 2))
```

```python
medical_data['Difference'] = dif_charges
print(medical_data)
```

```python
positive_values = medical_data[medical_data['Difference'] > 0]
print(positive_values)
```

```python
count = medical_data[medical_data['Difference'] > 0].count()
print(count)
```

```python
# Percentage of records with positive difference between actual charges and estimated costs

perc_pos_diff = (count / len(medical_data['Difference'])) * 100
print(round(perc_pos_diff, 2))
```

```python
# Percentage of negative differences

perc_neg_diff = 100 - perc_pos_diff
print(round(perc_neg_diff, 2))
```

## -- At this point, we can observe that almost 4.19 percent of the records have a positive difference between actual charges and estimated costs for Medical Insurance. 
Which means that, for this dataset, our estimated cost is overrated for 95.81 percent of the records (according with the data used for previous projects).

## -- There are some explanations for this situations, like:

    -- The actual calculation may use different weights associated with each variable, in relation to the calculation carried out to obtain the estimated costs;
    -- The actual cost may depend on other different prerequisites;
    -- The data is unbalanced at smoker status (smoker: no -> 1064 records and yes -> 274 records).
    -- In our estimated cost equation, be smoker or non-smoker has a huge weight in the calculations.
    -- Every region/state probably should have a different calculation formula for insurance cost.
